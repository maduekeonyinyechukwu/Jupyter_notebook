```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("all_recs.csv")
df.head()
```

```python
df = pd.read_csv("all_recs.csv", encoding='latin1')
df.head()
```

```python
year_order = df['Year'].value_counts().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Year")
sb.countplot(data=df, y='Year', order=year_order)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,6))
plt.title("Attack by Industry")
sb.countplot(data=df, y='Industry',order=industry_order)
plt.xticks(rotation=15);
```

```python
industry_order = df['Industry'].value_counts().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(10,6))
plt.title("Attack by Industry")
sb.countplot(data=df, y='Industry', order=industry_order)
plt.xlabel("Frequency")
plt.ylabel("Industry")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
